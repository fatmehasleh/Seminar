```python
# creating class of the issue
class Issuse( object ):  
    def __init__(self, repositoryurl, issuenumber, titleprocessed,bodyprocessed, label, labelcat, testtag, info,state, createdat, closedat ): 
        self.repositoryurl = repositoryurl
        self.issuenumber = issuenumber
        self.titleprocessed = titleprocessed
        self.bodyprocessed = bodyprocessed
        self.label = label
        self.labelcat = labelcat
        self.testtag = testtag
        self.info = info
        self.state = state
        self.createdat = createdat
        self.closedat = closedat

class RepositoryIssues( object ):  
    def __init__(self, repositoryurl, Issuse): 
        self.repositoryurl = repositoryurl
        self.issues = []

issue_s = []
repositoryIssues = []
```

```python
closed=[]
opend=[]
edges_bug=[]
edges_feature=[]
edges_support=[]
c=0
for ob in repositoryIssues:
  issueNumbers = []
  if c==1:
    for o in ob.issues:
      issueNumbers.append(o.issuenumber)
      issueNumbers.sort()
      if o.state == 'closed':
        closed.append(o)
      else: 
        print("titleprocessed= ",o.titleprocessed,"bodyprocessed= ", o.bodyprocessed,"label= ", o.label,"labelcat= ",o.labelcat,"testtag= ",o.testtag,"info=",o.info)
        opend.append(o)
    count=0
    for num in issueNumbers:
      count = count+1 
      for o in ob.issues:
        if num == o.issuenumber:
            issue_need = o
      if count %2 != 0:
        prev = issue_need
        if count>1: 
          if prev.labelcat == 'bug':
           edges_bug.append((prev2, prev))
          elif prev.labelcat == 'feature':
            edges_feature.append((prev2, prev))
          elif prev.labelcat == 'support':
            edges_support.append((prev2, prev))
      else: 
        if prev.labelcat == 'bug':
          edges_bug.append((prev, issue_need))
        elif prev.labelcat == 'feature':
          edges_feature.append((prev, issue_need))
        elif prev.labelcat == 'support':
          edges_support.append((prev, issue_need))
        prev2= issue_need 
    
  c=c+1
```

```python
val_map= {}

for f in edges_bug: 
  if f[0].state =='closed':
    val_map[f[0]] = 'red'
  if f[0].state=='open':
     val_map[f[0]] = 'green'
  if f[1].state =='closed':
    val_map[f[1]] = 'red'
  if f[1].state=='open':
     val_map[f[1]] = 'green'

for f in edges_feature: 
  if f[0].state =='closed':
    val_map[f[0]] = 'red'
  if f[0].state=='open':
     val_map[f[0]] = 'green'
  if f[1].state =='closed':
    val_map[f[1]] = 'red'
  if f[1].state=='open':
     val_map[f[1]] = 'green'

for f in edges_support: 
  if f[0].state =='closed':
    val_map[f[0]] = 'red'
  if f[0].state=='open':
     val_map[f[0]] = 'green'
  if f[1].state =='closed':
    val_map[f[1]] = 'red'
  if f[1].state=='open':
     val_map[f[1]] = 'green'
```

```python
val_map_edges= {}
for f in edges_bug: 
   val_map_edges[f[0]] = 2
for f in edges_feature: 
   val_map_edges[f[0]] = 2
for f in edges_support: 
   val_map_edges[f[0]] = 2
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_edges_from(edges_bug)
G.add_edges_from(edges_feature)
G.add_edges_from(edges_support)


values = [val_map.get(node, 'green') for node in G.nodes()]
values1= [val_map_edges.get(node, 1) for node in G.nodes()]

# Specify the edges you want here

purple_edges= edges_bug
tangerine_edges= edges_feature
beaver_edges=edges_support

plt.figure(figsize=(100,50))
# Need to create a layout when doing
# separate calls to draw nodes and edges
pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos, cmap=plt.get_cmap('jet'), node_color = values, node_size = 100 )
#nx.draw_networkx_labels(G, pos)
# nx.draw_networkx_edges(G, pos, edgelist=red_edges,width=values1, edge_color='red', arrows=True, arrowsize=20)
nx.draw_networkx_edges(G, pos, edgelist=purple_edges,width=values1, edge_color='purple', arrows=True, arrowsize=20)
nx.draw_networkx_edges(G, pos, edgelist=tangerine_edges,width=values1, edge_color='orange', arrows=True, arrowsize=20)
nx.draw_networkx_edges(G, pos, edgelist=beaver_edges,width=values1, edge_color='green', arrows=True, arrowsize=20)
# nx.draw_networkx_edges(G, pos, edgelist=yellow_edges, width=values1,edge_color='yellow', arrows=True, arrowsize=20)
# nx.draw_networkx_edges(G, pos, edgelist=edges_gpe, width=values1,edge_color='aquamarine', arrows=True, arrowsize=20)
# nx.draw_networkx_edges(G, pos, edgelist=edges_product, width=values1,edge_color='gray', arrows=True, arrowsize=20)
# nx.draw_networkx_edges(G, pos, edgelist=edges_art, width=values1,edge_color='brown', arrows=True, arrowsize=20)
# nx.draw_networkx_edges(G, pos, edgelist=black_edges, arrows=True, arrowsize=20)
plt.tight_layout()
plt.axis("off")
plt.show()
plt.savefig('foo.png')
```
